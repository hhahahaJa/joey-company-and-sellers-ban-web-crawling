```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import pandas as pd
import os

# 設定 Selenium 瀏覽器選項
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("--disable-notifications")

driver = webdriver.Chrome(options=options)

# 要爬取的網址列表
urls = [
        "https://sub.twincn.com/lm.aspx?q=22555003&sel=1",#統一超商
        "https://sub.twincn.com/lm.aspx?q=97165560&sel=1",#大潤發
        "https://sub.twincn.com/lm.aspx?q=29080592&sel=1",#楓康
        "https://sub.twincn.com/lm.aspx?q=03521905&sel=1",#遠百
        "https://sub.twincn.com/lm.aspx?q=97151664&sel=1",#寶雅
        "https://sub.twincn.com/lm.aspx?q=22853565&sel=1",#ok便利商店
        "https://sub.twincn.com/lm.aspx?q=53925190&sel=1",#Tomods 
        "https://sub.twincn.com/lm.aspx?q=55677060&sel=1",#松本清
        "https://sub.twincn.com/lm.aspx?q=84442769&sel=1",#大買家
        "https://sub.twincn.com/lm.aspx?q=53536349&sel=1",#日藥本舖
        "https://sub.twincn.com/lm.aspx?q=22555003&sel=1",#統一超商(好像沒跑到)
        "https://sub.twincn.com/lm.aspx?q=16740494&sel=1",#全聯
        "https://sub.twincn.com/lm.aspx?q=97165560&sel=1",#大潤發
        "https://sub.twincn.com/lm.aspx?q=23060248&sel=1",#全家
        "https://sub.twincn.com/lm.aspx?q=96972798&sel=1",#好事多
        "https://sub.twincn.com/lm.aspx?q=86649006&sel=1",#杏一
        "https://sub.twincn.com/lm.aspx?q=23285582&sel=1",#萊爾富
        "https://sub.twincn.com/lm.aspx?q=23224657&sel=1",#屈臣氏
        "https://sub.twincn.com/lm.aspx?q=80234156&sel=1",#佳瑪
        "https://sub.twincn.com/lm.aspx?q=22456427&sel=1",#東森
        "https://sub.twincn.com/lm.aspx?q=54164937&sel=1",#美聯社
        "https://sub.twincn.com/lm.aspx?q=89589328&sel=1",#康是美
        "https://sub.twincn.com/lm.aspx?q=16126526&sel=3",#新高橋藥局    
        #可以從台灣公司網找想要查詢的公司資料進行爬蟲
]

# 存放爬取結果
data = []

for url in urls:
    print(f"正在爬取: {url}")
    driver.get(url)
    time.sleep(3)#可調整等待頁面時時長
    while True:
        time.sleep(1)  
        rows = driver.find_elements(By.TAG_NAME, "tr")
        for row in rows:
            columns = row.find_elements(By.TAG_NAME, "td")
            
            if len(columns) >= 2:
                tax_id = columns[0].text.strip()
                try:
                    company_name_element = columns[1].find_element(By.TAG_NAME, "a")
                    company_name = company_name_element.text.strip()
                except:
                    company_name = "無法取得公司名稱"

                data.append([company_name, tax_id])  
                print([company_name, tax_id])

        # 嘗試尋找「下一頁」按鈕
        try:
            next_button = driver.find_element(By.LINK_TEXT, "下一頁")
            driver.execute_script("arguments[0].click();", next_button)
            time.sleep(3)
            print("跳轉至:", driver.current_url)
        except:
            break  # 沒有下一頁則跳出迴圈

# 關閉瀏覽器
driver.quit()
df = pd.DataFrame(data, columns=["公司名稱", "統一編號"])
# 存入 DataFrame
download_path = os.path.join(os.path.expanduser("~"), "Downloads", "統一編號資料1.xlsx")

# 將 DataFrame 存入 Excel
df.to_excel(download_path, index=False, engine="openpyxl")

print(f" 資料已存入 Excel 檔案：{download_path}")
```

```python
#台灣公司網輸入關鍵字獲取公司資料
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 設定搜尋關鍵字列表
search_keywords =lst # 這裡可以放多個搜尋詞

# 設定 Chrome 瀏覽器選項
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")
options.add_argument("--disable-blink-features=AutomationControlled")  # 隱藏 Selenium 特徵
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.6998.89 Safari/537.36")

# 啟動 WebDriver
driver = webdriver.Chrome(options=options)

# 存放所有搜尋結果的列表
all_tax_id_list = []
all_name_list = []
all_address_list = []
all_keyword_list = []  # 存放對應的關鍵字

# 遍歷所有關鍵字
for keyword in search_keywords:
    print(f"🔍 正在搜尋：{keyword}")

    driver.get("https://www.twincn.com/")
    time.sleep(2)  # 等待頁面加載

    # 找到輸入框並輸入關鍵字
    search_box = driver.find_element(By.ID, "ctl00_ContentPlaceHolder1_stxt")
    search_box.clear()  # 清空輸入框
    search_box.send_keys(keyword)
    search_box.send_keys(Keys.RETURN)
    time.sleep(3)  # 等待搜尋結果載入

    while True:
        time.sleep(2)  # 等待頁面內容載入
        rows = driver.find_elements(By.TAG_NAME, "tr")

        for row in rows:
            columns = row.find_elements(By.TAG_NAME, "td")

            if len(columns) >= 2:
                tax_id = columns[0].text.strip()  # 統一編號
                try:
                    company_name_element = columns[1].find_element(By.TAG_NAME, "a")
                    company_name = company_name_element.text.strip()
                except:
                    company_name = "無法取得公司名稱"

                try:
                    address_elements = columns[1].find_elements(By.CLASS_NAME, "col-sm-auto")
                    if len(address_elements) > 2:
                        address = address_elements[-2].text.strip()  # 擷取地址資訊
                    else:
                        address = "無法取得地址"
                except:
                    address = "無法取得地址"

                all_tax_id_list.append(tax_id)
                all_name_list.append(company_name)
                all_address_list.append(address)
                all_keyword_list.append(keyword)  # 記錄這筆資料的關鍵字

        # 嘗試尋找「下一頁」按鈕
        try:
            next_button = driver.find_element(By.LINK_TEXT, "下一頁")
            driver.execute_script("arguments[0].click();", next_button)
            time.sleep(3)  # 等待新頁面載入
            print(" 成功跳轉到:", driver.current_url)
        except:
            print(f" {keyword} 已經是最後一頁，結束爬取。")
            break

# 關閉瀏覽器
driver.quit()

# 建立 DataFrame 並顯示
data = {'搜尋關鍵字': all_keyword_list, '統一編號': all_tax_id_list, '店鋪名稱': all_name_list, '地址': all_address_list}
df = pd.DataFrame(data)

# 顯示 DataFrame
```
